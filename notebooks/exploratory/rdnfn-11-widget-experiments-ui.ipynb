```python
import ipywidgets as widgets

graphs = ["Polesia graph"]
habitats = ["Eagle habitat", "Lynx habitat"]
```

```python
graph_dropdown = widgets.Dropdown(
    options=graphs + habitats,
    description='',
)

view_label = widgets.Label("View:")

view_graph_btn = widgets.ToggleButton(
    value=False,
    description='Graph',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='View graph',
    icon=''
)
view_pgon_btn = widgets.ToggleButton(
    value=False,
    description='Polygons',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='View graph',
    icon=''
)
view_components_btn = widgets.ToggleButton(
    value=False,
    description='Components',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='View graph',
    icon=''
)

view_controls = widgets.HBox([view_label, view_graph_btn, view_pgon_btn, view_components_btn])
metrics = widgets.Label("Metrics go here.")
graph_accordion = widgets.Accordion(children=[metrics])
graph_accordion.set_title(0, 'Metrics')

graph_tab = widgets.VBox([graph_dropdown, view_controls, graph_accordion])
settings_tab = widgets.Label("Settings")

tab_nest = widgets.Tab()
tab_nest.children = [graph_tab, settings_tab]
tab_nest.set_title(0, 'Graphs')
tab_nest.set_title(1, 'Settings')
tab_nest
```
